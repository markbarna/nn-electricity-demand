```python
# plot load by region
```

```python
df.index = pd.to_datetime(df.index)
```

```python
ax = df['Load'].resample('D').mean().plot(figsize=(15,10))
ax.set_title('ERCOT Electricity Load')
ax.set_ylabel('Load')
yticks = ['{:,.0f} MW'.format(t) for t in ax.get_yticks()]
ax.set_yticklabels(yticks)
plt.show()
#plt.savefig('../Plots/Exploratory/monthly load')
```

```python
ax = df['Load'].groupby(df.index.hour).mean().plot(figsize=(15,10), xticks=range(23))
ax.set_title('ERCOT Mean Electricity Load (2017)')
ax.set_ylabel('Load')
ax.set_xlabel('Hour')
yticks = ['{:,.0f} MW'.format(t) for t in ax.get_yticks()]
ax.set_yticklabels(yticks)
xticks = ['{:0>2}:00'.format(t) for t in ax.get_xticks()]
ax.set_xticklabels(xticks, rotation='45')
plt.show()
#plt.savefig('../Plots/Exploratory/hourly load')
```

```python
# focus on coast region
```

```python
coast = df['COAST']
```

```python
coast = coast[['DryBulbTempC', 'Load']]
```

```python
coast['DryBulbTempCSq'] = coast['DryBulbTempC']**2
```

```python
coast.corr()
```

```python
coast[['Load', 'DryBulbTempC']].resample('D').mean().plot(figsize=(15,10), secondary_y='DryBulbTempC')
plt.show()
```

```python
coast[['Load', 'DryBulbTempCSq']].resample('D').mean().plot(figsize=(15,10), secondary_y='DryBulbTempCSq')
plt.show()
```

```python
coast.head()
```

```python
coast.diff(24)[['Load', 'DryBulbTempC']].resample('D').mean().plot(figsize=(15,10), secondary_y='DryBulbTempC')
plt.show()
```

```python
coast.diff(24)[['Load', 'DryBulbTempCSq']].resample('D').mean().plot(figsize=(15,10), secondary_y='DryBulbTempCSq')
plt.show()
```

```python
ax = coast.diff(24)['Load'].resample('D').mean().plot(figsize=(15,10))
ax.set_title('Differenced Electricity Load (Coast Region)')
ax.set_ylabel('Load')
yticks = ['{:,.0f} MW'.format(t) for t in ax.get_yticks()]
ax.set_yticklabels(yticks)
plt.show()
#plt.savefig('../Plots/Exploratory/diff monthly load')
```

```python
coast.diff(24)[['Load', 'DryBulbTempC']].groupby(df.index.hour).mean().plot(figsize=(15,10), secondary_y='DryBulbTempC')
plt.show()
```

```python
coast.diff(24)[['Load', 'DryBulbTempCSq']].groupby(df.index.hour).mean().plot(figsize=(15,10), secondary_y='DryBulbTempCSq')
plt.show()
```
